#Preentrega 1

El objetivo de esta preentega es examinar cómo las decisiones de política monetaria de la Reserva Federal de Estados Unidos afectan el mercado de acciones, específicamente a través del índice Standard and Poor 500 (S&P 500). Este índice comprende a las 500 compañías más grandes y ampliamente seguidas en los mercados financieros estadounidenses, proporcionando una representación de la situación mercado bursátil.

La Reserva Federal juega un papel crucial en la economía estadounidense, especialmente a través de sus ajustes en las tasas de interés. Reducir las tasas de interés, por ejemplo, puede impulsar la inversión y el gasto, lo que puede reflejar  aumentos en los precios de las acciones. Por el contrario, aumentos en las tasas de interés pueden desacelerar la economía y afectar negativamente las valoraciones del mercado de acciones.

En este breve análisis, nos enfocamos en un grupo dentro del S&P 500 conocido como el "Magnificent Seven", con foco en los precios de cierre. Estas empresas se destacan no solo por su tamaño y capitalización de mercado, sino también por su influencia significativa en el comportamiento general del índice.

Se analizarán también otras características del Magnificent Seven, tales como la volatilidad y el volumen operado. La volatilidad dará una idea de los riesgos de cada acción, mientras que el volumen operado reflejará el interés y la actividad de los inversores, obteniendo una visión más completa y detallada del desempeño de estas empresas.

Inicialmente, se utilizó un conjunto de datos amplio que abarcaba desde 2008 hasta 2023, posteriormente se restringió el análisis a las acciones del grupo, sumando la tasa de interés. El análisis visual de los datos se llevó a cabo mediante gráficos que abarcan el período y se destacaron tanto las caídas significativas como las subidas notables en los precios de las acciones del "Magnificent Seven".

#Gráfico 3

En este grafico se ven las acciones del Magnficent Seven junto a la tasa de interés de la Reserva Federal.

Para el primer período entre 2008 y 2010 se ve una baja drástica de la tasa de interés con el objetivo de minimizar los efectos negativos de la crisis de 2009, para maximizar la cantidad de capitales dirigidos hacia inversiones en cartera o en la economía real y no bajo la tasa de interés. La crisis vino acompañada de una caída del precio de las acciones, comenzando su recuperación una vez que la tasa de interés alcanzó valores mínimos.

Se comprende un segundo período de modificación de la tasa de interés entre 2016 y 2019 donde se sube la tasa de interés al mismo tiempo de la suba del precio de las acciones. Esto va en contra de lo que postula la teoría macroeconómica clásica pero hay algunas razones que lo explican.

En este período, la economía de Estados Unidos mostró un aumento significativo del PBI y una tasa de desempleo en descenso, elevando así los precios de las acciones. Además, mientras la Reserva Federal aumentaba las tasas de interés, otros bancos centrales mantuvieron políticas monetarias relajadas. Esto creó un entorno favorable, atrayendo inversiones extranjeras hacia las acciones estadounidenses. Otro factor clave fue la reforma fiscal aprobada en diciembre de 2017, que redujo significativamente la tasa de impuesto corporativo. Esta medida aumentó las ganancias de las empresas, lo que se reflejó positivamente en sus precios de acciones. Estos factores contribuyeron a que el aumento de las tasas de interés no resultara en una baja de los precios de las acciones.

La baja de tasa de interés entre 2019 y 2020 marcó un aumento en el precio de las acciones, a excepción de los primeros meses de confinamiento, que ya veníán creciendo en años anteriores.

La Reserva Federal bajó significativamente las tasas de interés en respuesta a la pandemia de COVID-19. Las tasas de interés más bajas redujeron el costo de los préstamos, estimulando el gasto y la inversión. Esto también hizo que las acciones fueran más atractivas en comparación con los bonos, cuyos rendimientos disminuyeron.
Además, los gobiernos implementaron paquetes de estímulo fiscal masivos para apoyar la economía. En Estados Unidos, esto incluyó cheques directos a los ciudadanos, ayudas a las pequeñas empresas y subsidios por desempleo, esto aumentó el dinero en circulación en la economía y mantuvo el consumo y la inversión en niveles decentes.

Por último, partir de 2022, las acciones bajaron debido a varias razones. La Reserva Federal incrementó agresivamente las tasas de interés para combatir la alta inflación, elevando los costos de endeudamiento y reduciendo el gasto y la inversión, sumado a la quita de ayudas monetarias implementadas en pandemia. La inflación persistente también aumentó los costos para las empresas y disminuyó el poder adquisitivo de los consumidores, disminuyendo la demanda de bienes y servicios. Además, las tensiones geopolíticas, como la guerra Ucrania-Rusia, crearon incertidumbre en los mercados globales, incrementando la incertidumbre

La conclusión a la que se puede llegar es que la tasa de interés tiene impacto en el precio de las accione al afectar el costo del capital y las decisiones de inversión. Al mismo tiempo, el resto de actores de la economía juegan su papel para mermar o potenfciar la fuerza de la medida.

#Gráfico 5

En este gráfico se observa la comparación del volumen operado del índice S&P 500 con la tasa de interés.
Se puede observar que en los períodos donde baja la tasa de interés aumenta el nivel de operaciones en las compañias dentro del índice, mientras que cuando sube la tasa de interés el volumen operado permanece invariable.

Una de las razones que explica este comportamiento es que cuando la tasa de interés es baja, se asocia con intentos de estimular la economía en períodos de desaceleración o recesión económica. Esto puede generar expectativas de recuperación económica y aumentar el optimismo del mercado, resultando en un mayor volumen operado. Mientras que una etapa de tasas de interés alta se suele implementar para controlar la inflación y pueden ser tomada como una señal de que la economía está sobrecalentada. Esto puede generar incertidumbre y hacer que los inversores sean más cautelosos, manteniendo constante el volumen operado.

```python
fig, eje1 = plt.subplots(figsize=(10, 6))

vol_op_mod["date"]= pd.to_datetime(vol_op_mod["date"])

eje1.bar(vol_op_mod["date"], vol_op_mod["volume"], label="Volumen Operado. Billones", color="blue", alpha=0.6)
eje1.set_xlabel("Fecha")
eje1.set_ylabel("Volumen Operado")

eje2 = eje1.twinx()
eje2.plot(sypint2023["Date"], sypint2023["Value"], label="Tasa de Interés", color="black")
eje2.set_ylabel("Tasa de Interés (%)")


fig.suptitle("Tasa de interés y Volumen operado en billones. Usar término en inglés, para claridad")
eje1.legend(loc="upper left")
eje2.legend(loc="upper right")

plt.show()
```
